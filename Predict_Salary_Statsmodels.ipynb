```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
data = {
    "Age": [25, 30, 35, 28, 32, 40, 45, 38],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "Finance", "IT", "HR"],
    "Salary": [50000, 60000, 70000, 55000, 65000, 80000, 75000, 68000]
}

df = pd.DataFrame(data)
df
```

```python
# Salary ~ Age + Department
model = smf.ols(formula="Salary ~ Age + Department", data=df).fit()
```

```python
print(model.summary())
```

```python
# Predict salary for a new employee
new_employee = pd.DataFrame({
    "Age": [29],
    "Department": ["IT"]
})

predicted_salary = model.predict(new_employee)
print("Predicted salary:", predicted_salary.iloc[0])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x="Age", y="Salary", hue="Department", data=df)
plt.title("Salary vs Age by Department")
plt.show()
```
