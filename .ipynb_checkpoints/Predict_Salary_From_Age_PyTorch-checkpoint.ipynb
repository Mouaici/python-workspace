```python
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = {
    "Age": [25, 30, 35, 28, 32],
    "Salary": [50000, 60000, 70000, 55000, 65000]
}

df = pd.DataFrame(data)
df
```

```python
# Input (Age)
X = torch.tensor(df['Age'].values, dtype=torch.float32).unsqueeze(1)  # shape [n_samples, 1]

# Target (Salary)
y = torch.tensor(df['Salary'].values, dtype=torch.float32).unsqueeze(1)
```

```python
class SalaryPredictor(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc = nn.Linear(1, 1)  # 1 input, 1 output
    
    def forward(self, x):
        return self.fc(x)

model = SalaryPredictor()
```

```python
criterion = nn.MSELoss()       # Mean Squared Error
optimizer = optim.SGD(model.parameters(), lr=0.001)
```

```python
epochs = 1000

for epoch in range(epochs):
    optimizer.zero_grad()
    y_pred = model(X)
    loss = criterion(y_pred, y)
    loss.backward()
    optimizer.step()
    
    if (epoch+1) % 200 == 0:
        print(f"Epoch {epoch+1}, Loss: {loss.item():.2f}")
```

```python
# Predict salary for a new employee: Age 40
new_age = torch.tensor([[40.0]])
predicted_salary = model(new_age).item()
print("Predicted salary for age 40:", round(predicted_salary, 2))
```

```python
plt.scatter(df['Age'], df['Salary'], color='blue', label='Actual')
plt.plot(df['Age'], model(X).detach().numpy(), color='red', label='Predicted')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Salary Prediction with PyTorch')
plt.legend()
plt.show()
```
