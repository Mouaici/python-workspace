```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
months = np.arange(1, 13)  # January to December
stores = ["Store_A", "Store_B", "Store_C"]

# Random monthly sales data
np.random.seed(42)
sales_data = np.random.randint(10000, 50000, size=(3, 12))  # 3 stores x 12 months
```

```python
ds = xr.DataArray(
    sales_data,
    coords=[stores, months],
    dims=["store", "month"],
    name="sales"
)

ds
```

```python
# Total sales per store
total_sales = ds.sum(dim="month")
print("Total sales per store:\n", total_sales)

# Average sales per month
avg_monthly_sales = ds.mean(dim="store")
print("Average monthly sales:\n", avg_monthly_sales)
```

```python
# Sales for Store_B
store_b_sales = ds.sel(store="Store_B")
store_b_sales
```

```python
# Sales for months 6 to 12
second_half = ds.sel(month=slice(6, 12))
second_half
```

```python
import matplotlib.pyplot as plt

# Convert to pandas Series for bar plot (simplest way)
total_sales.to_pandas().plot(kind="bar", title="Total Sales per Store")
plt.ylabel("Total Sales")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

for store in ds.store.values:
    plt.plot(ds.month.values, ds.sel(store=store).values, marker='o', label=store)

plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales per Store")
plt.legend()
plt.show()

```

```python
df = ds.to_pandas()  # shape: stores x months
df.T.plot(figsize=(8,5), marker='o')  # transpose so months are x-axis
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales per Store")
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Convert xarray to long-format DataFrame
df_long = ds.to_dataframe(name="Sales").reset_index()

plt.figure(figsize=(8,5))
sns.lineplot(x="month", y="Sales", hue="store", data=df_long, marker='o')
plt.title("Monthly Sales per Store")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```
