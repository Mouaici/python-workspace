```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("employees.csv")
df

```

```python
# Convert Salary column to NumPy array
salaries = df['Salary'].to_numpy()

# Compute mean, max, min
mean_salary = np.mean(salaries)
max_salary = np.max(salaries)
min_salary = np.min(salaries)

print("Mean salary:", mean_salary)
print("Max salary:", max_salary)
print("Min salary:", min_salary)

```

```python
# 10% raise, rounded to 2 decimals
df['Salary_Raised'] = np.round(df['Salary'] * 1.10, 2)
df
```

```python
# Average salary by department
dept_avg = df.groupby('Department')['Salary'].mean()
dept_avg

```

```python
plt.figure(figsize=(7,5))
plt.bar(dept_avg.index, dept_avg.values, color='skyblue')
plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Salary")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df['Name'], df['Salary'], marker='o', label='Original')
plt.plot(df['Name'], df['Salary_Raised'], marker='o', label='Raised')
plt.title("Salary Increase per Employee")
plt.xlabel("Employee")
plt.ylabel("Salary")
plt.legend()
plt.grid(True)
plt.show()
```

```python
df.to_csv('employees_final_jupyter.csv', index=False)
```
