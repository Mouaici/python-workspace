```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Eve", "Frank", "Grace", "Henry"],
    "Age": [25, 30, 35, 28, 32, 45, 41, 29],
    "Salary": [50000, 60000, 70000, 55000, 65000, 90000, 85000, 58000],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "Finance", "Finance", "IT"]
}

df = pd.DataFrame(data)
df
```

```python
sns.histplot(df["Salary"], kde=True)
plt.title("Salary Distribution")
plt.show()
```

```python
sns.boxplot(data=df, x="Department", y="Salary")
plt.title("Salary by Department")
plt.show()
```

```python
sns.scatterplot(data=df, x="Age", y="Salary", hue="Department", s=100)
plt.title("Age vs Salary by Department")
plt.show()
```

```python
sns.barplot(data=df, x="Department", y="Salary", estimator="mean")
plt.title("Average Salary by Department")
plt.show()
```
